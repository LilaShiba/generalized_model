```python
import stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/Users/kjams/Desktop/generalized_model/data/nyc.csv')
agent = stats.formulas(df)
```

```python
# knn
agent.init_knn(5, ['Black', 'Poverty'])
location = agent.scatterGraphNorm
testNode = agent.nodeList[200]
agent.insert_knn(testNode)
plt.scatter(location[0], location[1])
```

```python
x = agent.df['node'].tolist()
agent.knn_predict(x)
```

```python
# linear regression
x = agent.df['Unemployment'].tolist()
y = agent.df['Poverty'].tolist()
agent.linear_regression(x,y,40)
```

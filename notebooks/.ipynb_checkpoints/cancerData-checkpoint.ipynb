```python
import stats
import scipy.io
import numpy as np
import scipy as sp
import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mtx_file = sp.io.mmread('DLBCL1/matrix.mtx')
dense_matrix = mtx_file.toarray()
```

```python
genes = pd.read_csv("DLBCL1/genes.tsv", delimiter="\t", index_col=0, header=None, names=["gene_name"])
barcodes = pd.read_csv("DLBCL1/barcodes.tsv", delimiter="\t", index_col=0, header=None, names=["barcode"])
```

```python
df = pd.DataFrame(dense_matrix, index=genes.index, columns=barcodes.index)
```
